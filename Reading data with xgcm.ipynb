## Dataset Metadata
xarray makes it easy for us to clearly see the variables as well as the attributes of the dataset. Discussion on metadata can be found here: https://github.com/M6ASP/getting-started/blob/master/Reading%20data%20with%20scipy.io.netcdf.ipynb

```python
dataset = xarray.open_dataset('example_data/WOA13_annual_SST_nc4.nc')
dataset
```

We can also get individual variable data that shows the contents of the variable as well as its attributes:

```python
dataset['lat']
```

We will now assign variables for each of the variables.

```python
lat = dataset['lat']
lon = dataset['lon']
tos = dataset['tos']
```

We notice that there are quite a few `nan` values in `tos`. Therefore, we will have to perform a masking so those values are ignored.

```python
tos
```

Because sea surface temperature is irrelevant to our land masses, we will mask the hidden values and plot the map again so that the coloring excludes the continents.

```python
# helper function for masking arrays with missing values
def maskArray(handle):
    return numpy.ma.array(handle[:], mask=(handle[:] == nan))
```

```python
masked = maskArray(tos)
```

```python
plt.pcolormesh( masked )
plt.colorbar();

plt.title(tos.long_name);
plt.xlabel('%s (%s)' %(lon.long_name, lon.units));
plt.ylabel('%s (%s)' %(lat.long_name, lat.units));
```
